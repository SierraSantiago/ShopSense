```python
from selenium import webdriver
import pyautogui
import speech_recognition as sr
import azure.cognitiveservices.speech as speechsdk
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import WebDriverWait
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from datetime import datetime
import time
import pandas as pd
import re

website = 'https://www.mercadolibre.com/jms/mco/lgz/msl/login/H4sIAAAAAAAEAzWNPQ6DMAyF7-IZwZ6xF4lMMGDVISgxDRXi7nWqdvDwfr7nCyQtvHl97wQO6NyFAyt0sAvqnHL0PFkQxazCSj8ZUqtgxkhKuYC72tBC04MMalMzSiEr4aGrnyVV876_zOPi6TRuQ_GVxhdTS__Ekkysqntxw1Br7SPlgFMSHjP1IUW7Ae7OgKJeM4YnOM0H3R8hHuvEzgAAAA/user'
driver = webdriver.Chrome()
driver.get(website)
profileData = {'Item':[],'ItemInfo':[],'Distributor':[],'Link':[]}
time.sleep(65)
flag = True

while(flag):
    button = driver.find_element(By.XPATH, '//*[@id="app-root-wrapper"]/div/div/div[4]/nav/ul/li[6]/a/span')
    LinkData = driver.find_elements(By.XPATH, '//a[@class="list-item__link"]')
    DistributorData = driver.find_elements(By.XPATH, '//span[@class="text-with-link__text"]')
    ItemInfo = driver.find_elements(By.XPATH, '//p[@class="list-item__info"]')
    Item = driver.find_elements(By.XPATH, '//a//span[@class="bf-ui-rich-text"]')
    DateInfo = driver.find_elements(By.XPATH, '//span[@class="bf-ui-rich-text bf-ui-rich-text--bold"]')
    
    for index, element in enumerate(Item):
        if index % 2 == 0:
            if(element.text != "Ver mensajes"):
                profileData["Item"].append(element.text)
        
    for element in LinkData:
        if element.get_attribute('href') != 'https://articulo.mercadolibre.com.co/MCO-529025570-cable-original-cargador-iphone-5-6-7-8-x-lightning-1-metro-_JM':
            profileData["Link"].append(element.get_attribute("href"))
        
    for element in DistributorData:
        if (element.text != ''):
            if (element.text != 'ISHOP.STORE'):
                profileData["Distributor"].append(element.text)
    
    dates_to_exclude = ['3 de agosto de 2020', '9 de julio de 2020', '18 de junio de 2020', '16 de abril de 2020']

    for element, date in zip(ItemInfo, DateInfo):
        if date.text not in dates_to_exclude:
            profileData["ItemInfo"].append(element.text)
    
    driver.implicitly_wait(5)
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        button.click()
    
    except KeyboardInterrupt:
        flag = False
    except (StaleElementReferenceException, NoSuchElementException):
        flag = False
    except (TimeoutException, ElementNotInteractableException):
        flag = False  

#print(profileData["Item"])
#print(len(profileData["Item"]))

#print(profileData["ItemInfo"])
#print(len(profileData["ItemInfo"]))

#print(profileData["Distributor"])
#print(len(profileData["Distributor"]))

#print(profileData["Link"])
#print(len(profileData["Link"]))

driver.close()
```
